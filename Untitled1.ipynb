Converting object columns to integer values with LabelEncoder for easy visualisation when plotting

```python
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()
to_label = ['Job Title', 'Employment Type', 'Experience Level', 'Expertise Level', 'Company Location', 'Employee Residence', 'Company Size', 'Year']
for label in to_label:
    new_column_name = f"{label}_Labeled"
    practice[new_column_name] = lb.fit_transform(practice[label])
len(to_label)
```

```python
practice.head()
```

```python
newdf = practice.drop(['Job Title', 'Salary', 'Salary Currency', 'Employment Type', 'Experience Level', 'Expertise Level', 'Company Location', 'Employee Residence', 'Company Size', 'Year'], axis = 1)
```

```python
newdf.head()
```

```python
fig, axs = plt.subplots(ncols=9, nrows=1, figsize=(20, 3))
index = 0
axs = axs.flatten()
for k,v in newdf.items():
    sns.boxplot(y=k, data=newdf, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```
